# Steam Recommender System

For this assignment I chose to analyze the recommender system that Steam uses for video games. While they do have a web page, I actually used their desktop app as I was already signed in and familiar with the layout.

## 1. Scenario Analysis

### Who are your target users?

For Steam's recommender system, their target users are mainly video game players that use Steam's platform to play PC games. It could also expand to users looking for a game to gift to someone, though it would have a lot less data to go off of.

### What are their key goals?

The key goals of the target users include finding new games they will like with minimal effort on their end. They also might want to keep track of new and trendy games, as well as games their friends enjoyed.

### How can Steam help them accomplish those goals?

Steam can help users accomplish these goals by efficiently processing user data including play time, ratings/reviews, and user purchasing habits and trends, in order to curate an accurate list of games the user will enjoy. Steam seems to do a pretty good job of this already, having a robust interactive recommender, discovery queues, and very openly using a tag system in a user-facing way.

## 2. Reverse Engineering - How Does Steam's Recommender System Work?

The most obvious system that Steam uses in its recommendations is a tag system. Each game on the platform has its own list of tags that describe the game, ranging anywhere from genre to game mechanics to story content to the size of the development company and many more. This robust system is already a great start for any recommender system. Judging from data that Steam displays for the user, it seems that it also heavily involves play time, recent play time (last 2 weeks), whether you left a review, and finally whether you recommended the game or not (Steam's own thumbs up or thumbs down option). Steam also factors in the user's Friend List and can more heavily weigh what games are more popular among those users. This already seems like a lot of data to handle, and it is, but it seems that Steam utilizes one or more neural networks in order to process all of this data and make informed decisions.

## 3. Recommendations for Improvement

Honestly, Steam's current system is incredibly robust, accurate, and overall I personally enjoy using it. The sheer amount, and also breadth, of data that Steam uses to make their recommendations is impressive and results in a pretty precise and satisfying user experience. The front page immediately has a featured and recommended list that is easy and informative to scroll through. Steam will tell you what tags it's using to recommend any given game. As a user, you have access to a Discovery Queue: a tailored list of games that Steam thinks you might be interested in but would not seek out or find on your own. They also have two separate Ignore options, one that fully removes the game from the recommendation algorithm, and one that says you played it elsewhere so it can stop recommending this game, but its data stays in the algorithm if you enjoyed it. They also have an Interactive Recommender, where the user can control various aspects and variables, including sliders for looking at older vs. newer games, and the ability to filter by specific tags.

I think the biggest improvement that Steam could make is a bit of consolidation and simplification. The Discovery Queue, the Interactive Recommender, the Featured list, and many other recommended lists that pop up seemingly randomly all over the app are very spread out. If you're not super familiar with the layout, you might get lost in games that are being pushed only because they're on sale. I also think that tracking the play-time stat can be a little tricky. This feature can inadvertently skew toward longer games, not necessarily better ones, or ones with better replayability. Even the Playtime in the last 2 Weeks stat can potentially be misleading, giving recommendations a recency bias that might not be desired. A specific list or section of games related to ones you used to play would be very useful.
