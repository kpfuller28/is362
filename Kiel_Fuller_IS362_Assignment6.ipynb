```python
import pandas as pd

columns = ["MPG", "Cylinders", "Displacement", "Horsepower", "Weight", "Acceleration", "Model Year", "Origin", "Car Name"]
auto_df = pd.read_fwf("auto-mpg.data")
auto_df.columns = columns
```

```python
auto_df[auto_df['Horsepower'] == "?"]
```

```python
auto_df["Horsepower"] = auto_df["Horsepower"].replace(to_replace="?", value=0)
auto_df["Horsepower"] = pd.to_numeric(auto_df["Horsepower"])

auto_df.sort_values(by="Horsepower")
```

```python
auto_df["Origin"] = auto_df["Origin"].replace({1:"USA", 2:"Asia", 3:"Europe"})
auto_df.sort_values(by="Horsepower").head(10)
```

```python
auto_df['Cylinders'].value_counts().sort_index().plot.bar()
```

```python
auto_df.plot.scatter(x="Horsepower", y="Weight")
```

```python
import seaborn as sns

sns.lmplot(
    x="Weight",
    y="Acceleration",
    hue="Origin",
    col="Origin",
    data=auto_df,
    sharex=True,
    sharey=True
)
```
